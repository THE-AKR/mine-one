# Using z-score

```python
outliers=[]
def detect_outliers(a):
    threshold=3    #means 3rd standard deviation
    mean=np.mean(a)
    std=np.std(a)
    
    for i in a:
        z_score= (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_mpg= detect_outliers(mpg['mpg'])
outlier_mpg
```

```python
outlier_horse= detect_outliers(mpg["horsepower"])
outlier_horse
```
