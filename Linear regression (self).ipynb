```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
boston
```

```python
df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
df
```

```python
df.info()
```

```python
cor=df.corr
np.round(cor(),2)
```

```python
X=df    #independent variable 
y=boston.target    #dependent varibale or predictor
```

```python
y
```

```python
print(X.shape)
print(y.shape)
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#going for standardization , so it provide better result
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#hypertuning parameters
m=cross_val_score(regression,X_train,y_train,scoring="neg_mean_squared_error",cv=12)
```

```python
m
```

```python
np.mean(m) #closeness of data m with mean 
```

```python
pred=regression.predict(X_test)
```

```python
pred
```

```python
print(regression.score(X_test,pred))
print(regression.score(X_test,y_test))
```

```python
from sklearn.metrics import r2_score
score=r2_score(pred,y_test)
```

```python
score
```

```python
plt.scatter(y_test,pred)
plt.plot(y_test,regression.predict(),color='r')
```
